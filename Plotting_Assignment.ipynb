# **Carbon Dioxide Distribution**
This histogram below shows that largest amount of co2 was emmited from 250 to 550 and started slowing down.

```python
# Plot a histogram of carbon dioxide (CO2) column

plt.hist(data['co2'], bins=20)
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.title('Distribution of Carbon Dioxide (CO2)')
plt.show()
```

# **City Fuel Economy**
The histogram below shows city fuel economy distribution, roughly normal with a peak around 20. Most vehicles fall within the 15-25 range, with fewer vehicles showing extreme high or low fuel efficiency. Most vehicles have an average city fuel economy.

```python
# Plot a histogram of column 1
plt.hist(data['city'], bins=10)
plt.xlabel('City')
plt.ylabel('Frequency')
plt.show()
```

# **Engine Displacement**
The histogram shows engine displacement values with the x-axis representing displacement in liters and the y-axis showing frequency. Peaks indicate the most common engine sizes, while lower bars represent less common sizes. The shape of the histogram reveals the distribution pattern, helping analyze trends and correlations with factors like fuel efficiency and performance.

```python
# Plot a histogram of column 2
plt.hist(data['displ'], bins=15)
plt.xlabel('Engine Displacement')
plt.ylabel('Frequency')
plt.title('Distribution of Engine Displacement')
plt.show()
```

# **Correlation Heatmap**

```python
df = pd.DataFrame(data)

# Your original code
df_temp = df.select_dtypes(include=['float64', 'int64'])
corr_matric = df_temp.corr()
sns.heatmap(corr_matric, annot=False, cmap='coolwarm')
plt.show()
```

The diagram shows that engine displacement and the number of cylinders are negatively related to both fuel efficiency and greenhouse gas scores. Essentially, as engine size and cylinder count grow, vehicles generally become less efficient.

Additionally, engine displacement and cylinder count are positively linked with CO2 emissions, indicating that larger engines tend to release more CO2. There doesn’t appear to be any significant changes or trends in these variables over time.

## ****Correlation Interpretation:****

- **Strong Negative Correlation (Blue Color)**: Variables with values close to -1 reflect a strong inverse relationship, meaning as one variable increases, the other decreases.
- **Weak Negative Correlation (Light Blue Color)**: Features with values near 0 show a weaker inverse relationship, meaning the negative correlation exists, but is less pronounced.
- **Strong Positive Correlation (Red Color)**: Variables with values close to 1 represent a strong direct relationship, where both features increase together.
- **Weak Positive Correlation (Light Red Color)**: Values closer to 0 indicate a weak direct relationship, where the correlation is positive but not as significant as in darker red values.
